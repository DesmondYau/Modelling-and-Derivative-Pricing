### 2 Calibrating Hull White Model

#### 2.1 Instantaneous Forward Rate

The <b>instantaneous forward rate</b> at time t for investing at time T for a very short period is defined as:
$$f(t, T) = -\frac{\partial}{\partial T}log P(t,T)$$ 

```python
def f0t(t, curve):
    dt = 0.0001
    forward_rate = - (np.log(curve(t+dt)) - np.log(curve(t-dt))) / (2*dt)
    return forward_rate    
```

#### 2.2 Hull-White Short Rate Process

We will be modelling <b>short rate</b> using the <b>Hull-White model</b>, which is given by:
$$ \theta(t) = \frac{1}{\lambda}\frac{\partial}{\partial t}f(0,t) + f(0,t) + \frac{\sigma^2}{2\lambda^2}(1-e^{-2\lambda t})$$ 
$$dr(t) = \lambda (\theta (t) - r(t))dt + \sigma dW^{Q}(t)$$

Remarks:
- Hull White Model assumes <b>mean-reverting</b> and <b>normally distributed</b> interest rate
- $\lambda$ is the parameter for <b>speed of reversion</b>
- $\sigma$ is the parameter for <b>volatility</b>
- $\theta (t)$ is chosen so that the model <b>fits the term structure of interest rate currently observed in the market</b>. In fact, <b>the current forward rate curve f(0,t)</b> (i.e. forward rate at time 0 for investing at time t) is used to calibrate $\theta (t)$

```python
def generate_paths_HullWhite_euler(no_of_paths, no_of_steps, T, discount_curve, lambd, sigma):
    r0 = f0t(0.00001, discount_curve)
    theta = lambda t: 1.0/lambd * (f0t(t+dt, discount_curve) - f0t(t-dt, discount_curve))/(2*dt) + f0t(t, discount_curve) + (sigma**2)/(2*lambd**2)*(1-np.exp(-2*lambd*t))

    Z = np.random.normal(0.0, 1.0, [no_of_paths, no_of_steps])
    W = np.zeros([no_of_paths, no_of_steps+1])
    R = np.zeros([no_of_paths, no_of_steps+1])
    M = np.zeros([no_of_paths, no_of_steps+1])
    R[:,0] = r0
    M[:,0] = 1
    time = np.zeros([no_of_steps+1])

    dt = T / no_of_steps
    for i in range(0, no_of_steps):
        Z[:,i] = (Z[:,i] - np.mean(Z[:,i])) / np.std(Z[:,i])
        W[:,i+1] = W[:,i] + np.sqrt(dt) * Z[:,i]
        R[:,i+1] = R[:,i] + lambd * (theta(time[i]) - R[:,i]) * dt + sigma * (W[:,i+1] - W[:,i])
        M[:,i+1] = M[:,i] * np.exp((R[:,i+1] + R[:,i])/2 * dt)
        time[i+1] = time[i] + dt

    paths = {'Time': time, 'R': R, 'M':M}
        
        
    return paths
```

#### 2.3 Simulate Short Rate using Hull-White Model
- Note that in this exercise, we will only be using the Discount Curve from Market to perform calibration
- <b>Lambda parameter and Sigma parameter are specified and not estimated</b>

```python
no_of_paths = 2000                               # no. of paths for simulation
no_of_steps = 250                                # no. of steps for simulation
discount_curve = discount_curve_market           # market discount curve --> compute market forward rate curve --> compute theta Hull-White Model
lambd = 0.02                                     # lambda parameter in Hull-White Model
sigma = 0.02                                     # Sigma parameter in Hull-White Model
T = 40                                           # Perform simulation until maturity T

paths = generate_paths_HullWhite_euler(no_of_paths, no_of_steps, T, discount_curve, lambd, sigma)
ti = paths['Time']
rate = paths['R']

plt.figure(1)
plt.plot(ti, np.transpose(rate))   
plt.grid()
plt.xlabel("time")
plt.ylabel("Rate")
plt.title('Short Rate under Hull-White Model')
plt.xlabel('Time')
plt.ylabel('Short Rate')
```

### 2.4 Generating Discount Curve using Hull-White Model

Given short rate process r(t), prices of <b>Zero-Coupon Bond at time t for maturity T</b> is given by:

$$ B(t,T) = E^{Q}[e^{- \int_{t}^{T} r(s) \, ds}]$$

```python
M = paths['M']

discount_curve_HW = np.zeros([no_of_steps+1])
for i in range(0, no_of_steps+1):
    discount_curve_HW[i] = np.mean(1.0/M[:,i])
```

By comparing the <b>Market Discount Curve</b> vs <b>Hull White Discount Curve</b>, we show in this example that models under <b>HJM Framework</b>:
- incorporates <b>no arbitrage condition</b>
- fits to the <b>currently observed term structure in the market</b>

```python
plt.figure(1)
plt.grid()
plt.xlabel('Maturity (years)')
plt.ylabel('Discount Rate')
plt.plot(ti, discount_curve_market(ti))
plt.plot(ti, discount_curve_HW, '--r')
plt.legend(['Discount Curve - Market','Discount Curve - Hull White Model'])
plt.title('Market Discount Curve vs Hull White Discount Curve')
```

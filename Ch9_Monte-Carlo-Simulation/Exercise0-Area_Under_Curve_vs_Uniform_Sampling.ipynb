```python
import numpy as np
import matplotlib.pyplot as plt
```

### Summary
The idea of this exercise is to compare 2 approaches in Monte Carlo Simulation: <br>
- Compute integral of function g by <B>counting the number of samples under the curve<br>
- Compute integral of function g by <B>drawing uniform sample<br>

<br>
<br>

The following function would be used as an example:
$$\int_0^1 e^x \, dx = \left[ e^x \right]_0^1 = e - 1$$

<br>
<br>

#### 0. Setting Parameters

```python
no_of_samples = 10000
x1 = 0.0
x2 = 1.0
y1 = 0.0
y2 = 3.0

g = lambda x: np.exp(x)
result = g(1) - 1
print(f'The theoretical value is {result}')
```

#### 1. Compute integral of function g by counting the number of samples under the curve

```python
def compute_integral_area_under_curve(no_of_samples, x1, x2, y1, y2, g):
    x_i = np.random.uniform(x1, x2, no_of_samples)
    y_i = np.random.uniform(y1, y2, no_of_samples)

    proportion_under_curve = np.sum(g(x_i) > y_i)/no_of_samples                 # Calculating the proportion of dots under curve
    integral = proportion_under_curve * (y2 - y1) * (x2 - x1)
    
    
    plt.figure(1)
    plt.plot(x_i, y_i, '.r')
    plt.plot(x_i, g(x_i), '.b')
    return integral

result = compute_integral_area_under_curve(no_of_samples, x1, x2, y1, y2, g)
f'The simulated result is {result}'
```

#### 2. Compute integral of function g by drawing uniform sample

```python
def compute_integral_uniform_sampling(no_of_samples, x1, x2, g):
    x_i = np.random.uniform(x1, x2, no_of_samples)

    integral = np.mean(g(x_i)) * (x2 - x1)
    
    return integral

result = compute_integral_uniform_sampling(no_of_samples, x1, x2, g)
f'The simulated result is {result}'
```
